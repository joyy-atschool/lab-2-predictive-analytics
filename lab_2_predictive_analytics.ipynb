### Part 1.2 — Preprocessing & feature engineering
Clean the data and create features that help predict the tip.

````python
# TODO: Handle missing / invalid rows
#   e.g. drop or impute NaNs; remove rows where trip_distance <= 0 or fare_amount <= 0;
taxi = taxi.dropna()
taxi = taxi[(taxi['trip_distance'] > 0) &
        (taxi['fare_amount'] > 0)]

#   decide how to treat negative tip_amount values.
#There are no negative tip_amount values to treat.

# TODO: Feature engineering — create at least ONE new, justified feature, for example:
#   - tip_rate proxy, fare-per-mile = fare_amount / trip_distance
#   - total surcharges = mta_tax + tolls_amount + improvement_surcharge
#   (Careful: do NOT leak the target. Do not build features directly from tip_amount.)

taxi['total_surcharges'] = (
    taxi['mta_tax']
    + taxi['tolls_amount']
    + taxi['improvement_surcharge']
)

# TODO: Decide which columns are categorical (e.g. VendorID, RatecodeID, payment_type,
#   store_and_fwd_flag) vs numeric, and encode the categoricals (pd.get_dummies or OneHotEncoder).
categorical_cols = [
    'VendorID',
    'RatecodeID',
    'payment_type',
    'store_and_fwd_flag'
]

numeric_cols = [
    'passenger_count',
    'trip_distance',
    'fare_amount',
    'mta_tax',
    'tolls_amount',
    'improvement_surcharge',
    'total_surcharges'
]


taxi_encoded = pd.get_dummies(taxi, columns=categorical_cols, drop_first=True)

# TODO: Scale the numeric features.
#     (fit on TRAIN only — see Part 1.3)
from sklearn.preprocessing import StandardScaler


````

**Student Reasoning — Taxi preprocessing**
*1. How did you handle missing/invalid rows and why?*
*2. Which new feature(s) did you engineer and what is the intuition behind them?*
*3. Which scaling method did you choose and why is it appropriate here?*

> **Answer:**
1. I removed all rows containing missing values using dropna() as given, I also removed rows with negative trip distances and negative fare amounts, since they can negatively impact our model's learning process. Since no negative tip amounts were found , I didnt perform any actions on the tip_amount column.

2. A new feature which was engineered was total_surchages, and the formula was total surcharges = mta_tax + tolls_amount + improvement_surcharge. The intuition used is that the individual charges summed show the cost associated with a trip, which represent domain data. This domain data can be combined into one new feature alled total surchages which would indicate the overall trip cost and help improve how well the model makes predictions.

3. I used StandardScaler to scale the numeric features. StandardScaler is used in feature scaling of numeric features , where the values now have mean=0 and standard deviation = 1. This is appropriate here because the numerical columns are measured on different scales and their values may vary differently. Due to this, the model may give more improtance to the values which are higher, which is not valid. Scaling with this method allows the model to learn effectively by making all values have the same level of importance, as their values span the same range.


### Part 1.3 — Train / Validation / Test split
A three-way split lets you tune on validation data and keep the test set as a final, unbiased check.

````python
# TODO: Separate features (X) and target (y = tip_amount)
X = taxi_encoded.drop(columns=['tip_amount'])
y = taxi_encoded['tip_amount']


# TODO: Split into train / validation / test (a common choice is 60/20/20).
#
from sklearn.model_selection import train_test_split
#   First split off the test set, then split the remainder into train + validation.
#   Use random_state=RANDOM_STATE.
X_temp,X_test,y_temp,y_test = train_test_split(X,y,test_size = 0.2,random_state = RANDOM_STATE)

X_train,X_val,y_train,y_val = train_test_split(X_temp,y_temp,test_size = 0.25,random_state = RANDOM_STATE)
# TODO: Fit your scaler on the TRAINING set only, then transform train, val, and test.
#   (Explain in the reasoning box why scaling must be fit on train only.) please refer to markdown.
s = StandardScaler()

X_train[numeric_cols] = s.fit_transform(X_train[numeric_cols])

X_val[numeric_cols] = s.transform(X_val[numeric_cols])

X_test[numeric_cols] = s.transform(X_test[numeric_cols])
````

**Student Reasoning — Splitting**
*What ratio did you use? Why is a separate validation set useful in addition to a test set?
Why must the scaler (and any imputation statistics) be fit on the training data only?*

> **Answer:**

I used 60/20/20 as you suggested. 60% for training, 20% for validation and the last 20% for testing. The separate validation test is useful in selecting the best model without touching the final test data , which the model could memorise(overfitting). It also prevents data leakage, by ensuring that the test set is hideen from the model for optimal performance. It is also useful for something called hyperparameter tuning, where we repeatedly use the validation set to adjust model setting to improve model performance.

The scaler must be fit on the training data to mainly prevent data leakage. If we use validation or test sets, the model is going to directly learn what it is going to be evaluated on, and this would not show whether true learning or improvement occured. Eventually, this would lead to overly positive performance from the model, and would give an untrue reflection of how the model would perform in the real world.

### Part 1.4 — Train a regressor and check for overfitting
Train a model and evaluate it on **train, validation, and test** sets.

````python
# TODO: Import, initialise, and train a regression model.
#   Start simple (LinearRegression), then try a more flexible model
#   (e.g. RandomForestRegressor or DecisionTreeRegressor) and compare.
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# TODO: Predict on train, validation, and test sets.
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)


y_train_linear_pred = linear_model.predict(X_train)
y_val_linear_pred = linear_model.predict(X_val)
y_test_linear_pred= linear_model.predict(X_test)


# TODO: Report RMSE and R^2 for ALL THREE sets.
#
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Train
train_rmse_linear = mean_squared_error(y_train, y_train_linear_pred) ** 0.5
train_r2_linear = r2_score(y_train, y_train_linear_pred)

# Validation
val_rmse_linear = mean_squared_error(y_val, y_val_linear_pred) ** 0.5
val_r2_linear = r2_score(y_val, y_val_linear_pred)

# Test
test_rmse_linear = mean_squared_error(y_test, y_test_linear_pred) ** 0.5
test_r2_linear = r2_score(y_test, y_test_linear_pred)
#   (RMSE = mean_squared_error(y_true, y_pred) ** 0.5)
print("Linear Regression")
print(f"Train RMSE: {train_rmse_linear:.3f}, R²: {train_r2_linear:.3f}")
print(f"Validation RMSE: {val_rmse_linear:.3f}, R²: {val_r2_linear:.3f}")
print(f"Test RMSE: {test_rmse_linear:.3f}, R²: {test_r2_linear:.3f}")


#Random Forest
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=RANDOM_STATE
)

rf_model.fit(X_train, y_train)

y_train_pred_rf = rf_model.predict(X_train)
y_val_pred_rf = rf_model.predict(X_val)
y_test_pred_rf = rf_model.predict(X_test)

train_rmse_rf = mean_squared_error(y_train, y_train_pred_rf) ** 0.5
train_r2_rf = r2_score(y_train, y_train_pred_rf)

val_rmse_rf = mean_squared_error(y_val, y_val_pred_rf) ** 0.5
val_r2_rf = r2_score(y_val, y_val_pred_rf)

test_rmse_rf = mean_squared_error(y_test, y_test_pred_rf) ** 0.5
test_r2_rf = r2_score(y_test, y_test_pred_rf)

print("\nRandom Forest")
print(f"Train RMSE: {train_rmse_rf:.3f}, R²: {train_r2_rf:.3f}")
print(f"Validation RMSE: {val_rmse_rf:.3f}, R²: {val_r2_rf:.3f}")
print(f"Test RMSE: {test_rmse_rf:.3f}, R²: {test_r2_rf:.3f}")

````
